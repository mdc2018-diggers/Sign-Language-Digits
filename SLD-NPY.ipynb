```python
import tensorflow as tf
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import time
import datetime
import math
from sklearn.metrics import confusion_matrix
from datetime import timedelta
from sklearn.model_selection import train_test_split
%matplotlib inline

import glob
import os
import operator
import functools
import random
from scipy.ndimage.interpolation import rotate, shift, zoom
```

```python
def get_label(index):
    correct_label = [9, 0, 7, 6, 1, 8, 4, 3, 2, 5]
    return correct_label[index]

def get_labels(indexes):
    return [get_label(index) for index in indexes]
```

```python
# load data set
all_X = np.load('./Datasets/npy_dataset/X.npy')
all_Y = np.load('./Datasets/npy_dataset/Y.npy')
```

```python
r = range(200,209)
# Get the first images from the test-set.
images = all_X[r]

# Get the true classes for those images.
cls_true = get_labels(np.argmax(all_Y, axis=1)[r])

# Plot the images and labels using our helper-function above.
plot_images(images=images, cls_true=cls_true)
```

```python
# Train Test split
test_size = 0.15
X_train, X_test, Y_train, Y_test = train_test_split(all_X, all_Y, 
                                                    test_size=test_size, 
                                                    random_state=42)
```

```python
# We know that MNIST images are 28 pixels in each dimension.
img_size = 64

# Images are stored in one-dimensional arrays of this length.
img_size_flat = img_size * img_size

# Tuple with height and width of images used to reshape arrays.
img_shape = (img_size, img_size)

# Number of colour channels for the images: 1 channel for gray-scale.
num_channels = 1

# Number of classes, one class for each of 10 digits.
num_classes = 10
```

```python
def plot_images(images, cls_true, cls_pred=None):
    assert len(images) == len(cls_true) == 9
    
    # Create figure with 3x3 sub-plots.
    fig, axes = plt.subplots(3, 3)
    fig.subplots_adjust(hspace=0.3, wspace=0.3)

    for i, ax in enumerate(axes.flat):
        # Plot image.
        ax.imshow(images[i].reshape(img_shape), cmap='binary')

        # Show true and predicted classes.
        if cls_pred is None:
            xlabel = "True: {0}".format(cls_true[i])
        else:
            xlabel = "True: {0}, Pred: {1}".format(cls_true[i], cls_pred[i])

        # Show the classes as the label on the x-axis.
        ax.set_xlabel(xlabel)
        
        # Remove ticks from the plot.
        ax.set_xticks([])
        ax.set_yticks([])
    
    plt.show()
```

```python
# Get the first images from the test-set.
images = X_test[0:9]

# Get the true classes for those images.
cls_true = Y_test_cls[0:9]

# Plot the images and labels using our helper-function above.
plot_images(images=images, cls_true=cls_true)
```
